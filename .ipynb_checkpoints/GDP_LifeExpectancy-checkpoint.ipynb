### Descriptive Analysis

```python
# What is the average life expectancy across all countries each year?
le_mean_each_year = dataset.groupby('year')[['le_years']].mean().reset_index()
le_mean_each_year
```

```python
# What is the average gdp across all countries each year?
gdp_mean_each_year = dataset.groupby('year')[['gdp']].mean().reset_index()
gdp_mean_each_year
```

```python
# year ticks and labels
year_ticks = list(range(0, 16))

year_labels = []
for i in year_ticks:
    year_labels.append(str(i))
```

```python
year_labels
```

```python
# Data visualization:

plt.figure(figsize=(10, 5))

# life expectancy subplot
ax_life_expectancy = plt.subplot(1, 2, 1)
sns.barplot(data=le_mean_each_year, x='year', y='le_years', color=(0.3, 0.3, 0.255, 0.74))

ax_life_expectancy.set_xticks(year_ticks)
ax_life_expectancy.set_xticklabels(year_labels)


plt.title("Average Life Expectancy per Year")
plt.xlabel('Years (2000s)')
plt.ylabel("Life Expectancy at birth (Years)")


# gdp subplot
ax_gdp = plt.subplot(1, 2, 2)
sns.barplot(data=gdp_mean_each_year, x='year', y='gdp', color=(0.3, 0.7, 0.255, 0.74))

ax_gdp.set_xticks(year_ticks)
ax_gdp.set_xticklabels(year_labels)


plt.title("Average GDP per Year")
plt.xlabel('Years (2000s)')
plt.ylabel("GDP (trillions)")




plt.show()
plt.close()
```

Average life expectancy across all countries increased slightly every year; while GDP increased drastically throughout the span of 15 years.

```python
# which country had the highest and lowest life expectancy each year

sns.scatterplot(data=dataset, x='year', y='le_years', hue='country')

plt.legend(bbox_to_anchor=(1.1, 1.05))

plt.xticks(range(2000, 2016), year_labels)

plt.title("Yearly Life Expectancy for each Country")
plt.xlabel('Years (2000s)')
plt.ylabel("Life Expectancy at Birth (years)")

plt.show()
plt.close()
```

- Germany has the highest life expectancy every year.
- Zimbabwe has the lowest life expectancy every year.

```python
# which country had the highest and lowest gdp each year

sns.scatterplot(data=dataset, x='year', y='gdp', hue='country')

plt.legend(bbox_to_anchor=(1.1, 1.05))

plt.xticks(range(2000, 2016), year_labels)

plt.title("Yearly GDP (in trillions $)")
plt.xlabel('Years (2000s)')
plt.ylabel("GDP (in trillions $)")

plt.show()
plt.close()
```

- The US has the highest GDP from years 2000-2015.
- Zimbabwe had lowest GDP from years 2000-2015, but comparable to Chile.
- Countries US, Germany, and Mexico had a dip in GDP from 2008 to 2009, while China seems to be unaffected or did not dip.

### Trend Over Time

```python
# How has the life expectancy at birth changed over time for each country?

# get the first and last years rows for each country
first_last_years = dataset[(dataset['year'] == 2000) | (dataset['year'] == 2015)]

# new dataframe for differences in life expectancy
start_to_end_le_years = pd.merge(first_last_years.groupby('country')[['le_years']].min().reset_index(),
                               first_last_years.groupby('country')[['le_years']].max().reset_index(),
                               on='country', suffixes=('_min', '_max'))

start_to_end_le_years['difference'] = start_to_end_le_years.le_years_max - start_to_end_le_years.le_years_min
```

```python
start_to_end_le_years
```

```python
# list of bounds
bottom_years = start_to_end_le_years.le_years_min.to_list()
upper_years = start_to_end_le_years.le_years_max.to_list()
```

```python
plt.bar(x=range(len(start_to_end_le_years)), height=start_to_end_le_years.le_years_max - start_to_end_le_years.le_years_min,
        bottom=start_to_end_le_years.le_years_min)

plt.ylim(40, 90)

plt.yticks(range(40, 90, 5))

plt.xticks(range(6), ['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'])

plt.title("15 Year Spread of Life Expectancy")
plt.xlabel('Country')
plt.ylabel('Life Expectancy at birth (in years)')


plt.show()
plt.close()
```

- Zimbabwe had the highest increase of Life Expectancy in the span of 15 years, while Mexico had the least increase.
- Germany stands at the highest of life expectancy at birth (in years), while Zimbabwe stands at the lowest despite being the country with the highest increase in the 15 year period.

```python
# How has the GDP changed over time for each country?

# get the first and last years rows for each country
first_last_years = dataset[(dataset['year'] == 2000) | (dataset['year'] == 2015)]

# new dataframe for differences in life expectancy
start_to_end_gdp = pd.merge(first_last_years.groupby('country')[['gdp']].min().reset_index(),
                               first_last_years.groupby('country')[['gdp']].max().reset_index(),
                               on='country', suffixes=('_min', '_max'))

start_to_end_gdp['difference'] = start_to_end_gdp.gdp_max - start_to_end_gdp.gdp_min
```

```python
# list of bounds
bottom_gdp = start_to_end_gdp.gdp_min.to_list()
upper_gdp = start_to_end_gdp.gdp_max.to_list()
```

```python
plt.bar(x=range(len(start_to_end_gdp)), height=start_to_end_gdp.gdp_max - start_to_end_gdp.gdp_min,
        bottom=start_to_end_gdp.gdp_min)

plt.xticks(range(6), ['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'])

plt.title("15 Year Spread of GDP")
plt.xlabel('Country')
plt.ylabel('GDP')


plt.show()
plt.close()
```

- Although the USA has the highest GDP in 2015, China had the highest increase in the span of 15 years compared to all the other countries. 
- Zimbabwe had the lowest change in GDP.

### Correlations and Relationships

```python
# scatterplot GDP and Life Expectancy of all countries altogether

sns.scatterplot(data=dataset, x='le_years', y='gdp', hue='country')

plt.show()
plt.close()
```

```python
# by country scatter plots

plt.figure(figsize=(10, 20))

plt.subplots_adjust(right=1, bottom=0.2)

# Chile
ax_chile = plt.subplot(3, 2, 1)
sns.scatterplot(data=dataset[dataset['country'] == 'Chile'], x='le_years', y='gdp')

plt.title("Chile")
plt.xlabel('Life Expectancy at Birth (in years)')
plt.ylabel("GDP")

# China
ax_china = plt.subplot(3, 2, 2)
sns.scatterplot(data=dataset[dataset['country'] == 'China'], x='le_years', y='gdp')

plt.title("China")
plt.xlabel('Life Expectancy at Birth (in years)')
plt.ylabel("GDP")

# Germany
ax_germany = plt.subplot(3, 2, 3)
sns.scatterplot(data=dataset[dataset['country'] == 'Germany'], x='le_years', y='gdp')

plt.title("Germany")
plt.xlabel('Life Expectancy at Birth (in years)')
plt.ylabel("GDP")

# Mexico
ax_mexico = plt.subplot(3, 2, 4)
sns.scatterplot(data=dataset[dataset['country'] == 'Mexico'], x='le_years', y='gdp')

plt.title("Mexico")
plt.xlabel('Life Expectancy at Birth (in years)')
plt.ylabel("GDP")

# United States of America
ax_usa = plt.subplot(3, 2, 5)
sns.scatterplot(data=dataset[dataset['country'] == 'United States of America'], x='le_years', y='gdp')

plt.title("USA")
plt.xlabel('Life Expectancy at Birth (in years)')
plt.ylabel("GDP")

# Zimbabwe
ax_zimbabwe = plt.subplot(3, 2, 6)
sns.scatterplot(data=dataset[dataset['country'] == 'Zimbabwe'], x='le_years', y='gdp')

plt.title("Zimbabwe")
plt.xlabel('Life Expectancy at Birth (in years)')
plt.ylabel("GDP")


plt.show()
plt.close()
```

- There is an overall positive correlation between Life expectancy and GDP.
